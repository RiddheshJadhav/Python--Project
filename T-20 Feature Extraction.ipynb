```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pickle.load(open('dataset_level2.pkl','rb'))
```

```python
df

# batting_team
# bowling team
# city
# current score
# ball left
# wickets left
# current rr
# last five
```

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()  # name of the venues where the city data is null
```

```python
cities = np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city']) 

# Getting the names of the cities with the help of venue by spiling its first word
```

```python
df['city'] = cities
```

```python
df.isnull().sum()  # No more null values present in the cities
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df
```

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()

# selecting only the cities where at least 5 matches are played i.e 600 bowls are bowled
```

```python
df = df[df['city'].isin(eligible_cities)]  # Getting the data only the data of cities with at least 5 matches
```

```python
df
```

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']   # Gettting current score till the ball
```

```python
df
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split(".")[1])
df

# Getting the data for the over number and bowl number from ball column by splting it
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
df

# Calculating the number of bowls bowled
```

```python
df['balls_left'] = 120 - df['balls_bowled']   # Calculating the no of balls left in innings
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)  # comdition when balls left becomes -ve i.e extra ball are bowled
df
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)   # calculating the wickets fallen where name of player is diplayed
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left'] = 10 - df['player_dismissed']  
```

```python
df
```

```python
df['crr'] = (df['current_score']*6)/df['balls_bowled']   # Calcualting the current run rate
df
```

```python
groups = df.groupby('match_id')  

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    last_five.extend(groups.get_group(id).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five'] = last_five
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id') 
# Getting the total runs scored
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five','runs_x']]
```

```python
final_df
```

```python
final_df.dropna(inplace=True)
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
X = final_df.drop(columns=['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
